```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df
```

## Player Count

* Display the total number of players


```python
total_players = purchase_data_df["SN"].value_counts(dropna=False)
total = len(total_players)
print(f"Total number of players = {total}")
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data_df["Item Name"].value_counts(dropna=False)
num_unique_items = len(unique_items)

avg_price = purchase_data_df["Price"].mean()
tot_price = purchase_data_df["Price"].sum()

purchases = purchase_data_df["Purchase ID"].value_counts()
num_purchases = len(purchases)

summary_dict = [{
    "Number of Unique Items" : num_unique_items,
    "Average Price" : round(avg_price, 2), 
    "Number of Purchases" : num_purchases,
    "Total Revenue" : round(tot_price, 2)
}]

summary_df = pd.DataFrame(summary_dict)

summary_df["Average Price"] = summary_df["Average Price"].map("${:.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:.2f}".format)

summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
newdata4_df = purchase_data_df[["SN", "Age"]]
uniqueSNdata4_df = pd.DataFrame.drop_duplicates(newdata4_df)

bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]
age_groups = ["<10", "10 - 14", "15 - 19", "20 - 24", "25 - 29", "30 - 34", "35 - 39", "40 +"]

uniqueSNdata4_df["Age Group"] = pd.cut(uniqueSNdata4_df["Age"], bins, labels=age_groups, include_lowest=True)

unique5 = uniqueSNdata4_df.groupby("Age Group")

total_count = uniqueSNdata4_df["SN"].count()

age_count = unique5["SN"].count()



age_demographics_df = pd.DataFrame({"Total Count" : age_count})

age_percent = round(age_demographics_df["Total Count"] / total_count *100, 2)

age_demographics_df["Percentage of Players"] = age_percent

age_demographics_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
newdata5_df = purchase_data_df[["SN", "Price", "Purchase ID"]]
SNregroup = newdata5_df.groupby("SN")

purchase_count = SNregroup["Purchase ID"].count()

total_purchase = SNregroup["Price"].sum()

avg_purchase = round(SNregroup["Price"].mean(), 2)

spenders_df = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Average Purchase": avg_purchase,
    "Total Purchase": total_purchase
})

topspenders_df = spenders_df.sort_values("Total Purchase", ascending = False)

topspenders_df["Average Purchase"] = topspenders_df["Average Purchase"].map("${:.2f}".format)
topspenders_df["Total Purchase"] = topspenders_df["Total Purchase"].map("${:.2f}".format)

topspenders_df.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
newdata6_df = purchase_data_df[["Item ID", "Item Name", "Price"]]
itemregroup = newdata6_df.groupby(["Item ID", "Item Name"])


purchase_count = itemregroup["Item ID"].count()

total_purchase = itemregroup["Price"].sum()

avg_purchase = round(itemregroup["Price"].mean(), 2)

items_df = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Item Price": avg_purchase,
    "Total Purchase Value": total_purchase
})

items_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
topitems_df = items_df.sort_values("Total Purchase Value", ascending = False)

topitems_df["Item Price"] = topitems_df["Item Price"].map("${:.2f}".format)
topitems_df["Total Purchase Value"] = topitems_df["Total Purchase Value"].map("${:.2f}".format)
topitems_df.head(5)
```
